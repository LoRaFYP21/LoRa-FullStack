# PDR Analysis

# PDR SF = 7

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read PDR CSV file
pdr_df = pd.read_csv('PDR.csv')

# Clean the data - replace 'X' with NaN and convert to numeric
for col in ['Test 1 (%)', 'Test 2 (%)', 'Test 3 (%)', 'Avg (%)']:
    pdr_df[col] = pdr_df[col].replace('X', np.nan)
    pdr_df[col] = pd.to_numeric(pdr_df[col], errors='coerce')

# Filter data for SF 7
pdr_sf7_data = pdr_df[pdr_df['SF'] == 7].copy()

# Graph 1: Grouped Bar Chart - PDR for Each Fragment Size (All Tests)
plt.figure(figsize=(12, 6))

# Set up data for grouped bars
fragment_sizes = pdr_sf7_data['Fragment Size (B)'].astype(str)
test1_pdr = pdr_sf7_data['Test 1 (%)']
test2_pdr = pdr_sf7_data['Test 2 (%)']
test3_pdr = pdr_sf7_data['Test 3 (%)']
avg_pdr = pdr_sf7_data['Avg (%)']

# Set the width of bars and positions
bar_width = 0.2
x_pos = np.arange(len(fragment_sizes))

# Create grouped bars
bars1 = plt.bar(x_pos - 1.5*bar_width, test1_pdr, bar_width, 
                label='Test 1', color='#E74C3C', alpha=0.8, edgecolor='black')
bars2 = plt.bar(x_pos - 0.5*bar_width, test2_pdr, bar_width, 
                label='Test 2', color='#3498DB', alpha=0.8, edgecolor='black')
bars3 = plt.bar(x_pos + 0.5*bar_width, test3_pdr, bar_width, 
                label='Test 3', color='#F39C12', alpha=0.8, edgecolor='black')
bars_avg = plt.bar(x_pos + 1.5*bar_width, avg_pdr, bar_width, 
                   label='Average', color='#27AE60', alpha=0.8, edgecolor='black', linewidth=2)

plt.xlabel('Fragment Size (B)', fontsize=14, fontweight='bold')
plt.ylabel('Packet Delivery Ratio (%)', fontsize=14, fontweight='bold')
plt.title('PDR SF 7: Packet Delivery Ratio by Fragment Size (All Tests)', fontsize=16, fontweight='bold')
plt.xticks(x_pos, fragment_sizes)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3, axis='y', linestyle='--')
plt.ylim(95, 101)  # Set y-axis limits for better visibility of differences

# Add value labels on all bars
for bars in [bars1, bars2, bars3, bars_avg]:
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 f'{height:.1f}',
                 ha='center', va='bottom', fontsize=7, fontweight='bold')

plt.tight_layout()
plt.show()

# Graph 2: Line Chart - Variation of Average PDR with Fragment Size
plt.figure(figsize=(10, 6))
plt.plot(pdr_sf7_data['Fragment Size (B)'], pdr_sf7_data['Avg (%)'], 
         marker='o', linewidth=3, markersize=10, color='#2E86AB')

plt.xlabel('Fragment Size (B)', fontsize=14, fontweight='bold')
plt.ylabel('Average Packet Delivery Ratio (%)', fontsize=14, fontweight='bold')
plt.title('PDR SF 7: Average Packet Delivery Ratio vs Fragment Size', fontsize=16, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle='--')
plt.xticks(pdr_sf7_data['Fragment Size (B)'])
plt.ylim(99, 100.5)  # Set y-axis limits for better visibility

# Add value labels on points
for i, (x, y) in enumerate(zip(pdr_sf7_data['Fragment Size (B)'], pdr_sf7_data['Avg (%)'])):
    plt.annotate(f'{y:.2f}%', (x, y), textcoords="offset points", 
                xytext=(0,10), ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

# Combined Plot: PDR for SF 7, 8, 9, 10

```python
# Combined plot for SF 7, 8, 9, and 10
plt.figure(figsize=(14, 8))

# Define colors for each SF
colors = {'SF 7': '#E74C3C', 'SF 8': '#3498DB', 'SF 9': '#F39C12', 'SF 10': '#27AE60'}
markers = {'SF 7': 'o', 'SF 8': 's', 'SF 9': '^', 'SF 10': 'D'}

# Plot data for each SF
for sf in [7, 8, 9, 10]:
    sf_data = pdr_df[pdr_df['SF'] == sf].copy()
    plt.plot(sf_data['Fragment Size (B)'], sf_data['Avg (%)'], 
             marker=markers[f'SF {sf}'], linewidth=3, markersize=8, 
             color=colors[f'SF {sf}'], label=f'SF {sf}', alpha=0.8)

plt.xlabel('Fragment Size (B)', fontsize=16, fontweight='bold')
plt.ylabel('Average Packet Delivery Ratio (%)', fontsize=16, fontweight='bold')
plt.title('PDR Comparison: Average Packet Delivery Ratio vs Fragment Size\nfor SF 7, 8, 9, and 10', 
          fontsize=18, fontweight='bold', pad=20)
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=14, loc='best', frameon=True, shadow=True)

# Set x-axis ticks to show all fragment sizes
all_fragment_sizes = sorted(pdr_df['Fragment Size (B)'].unique())
plt.xticks(all_fragment_sizes, rotation=45)

# Set y-axis limits to show the full range of data
plt.ylim(94, 101)

# Add annotations for key data points if needed
for sf in [7, 8, 9, 10]:
    sf_data = pdr_df[pdr_df['SF'] == sf].copy()
    # Annotate the minimum and maximum points for each SF
    min_idx = sf_data['Avg (%)'].idxmin()
    max_idx = sf_data['Avg (%)'].idxmax()
    
    min_x = sf_data.loc[min_idx, 'Fragment Size (B)']
    min_y = sf_data.loc[min_idx, 'Avg (%)']
    max_x = sf_data.loc[max_idx, 'Fragment Size (B)']
    max_y = sf_data.loc[max_idx, 'Avg (%)']
    
    # Only annotate if there's significant variation
    if max_y - min_y > 0.5:
        plt.annotate(f'Min: {min_y:.1f}%', (min_x, min_y), 
                    textcoords="offset points", xytext=(10,10), 
                    ha='left', fontsize=9, color=colors[f'SF {sf}'], fontweight='bold')
        plt.annotate(f'Max: {max_y:.1f}%', (max_x, max_y), 
                    textcoords="offset points", xytext=(10,-15), 
                    ha='left', fontsize=9, color=colors[f'SF {sf}'], fontweight='bold')

plt.tight_layout()
plt.show()

# Create a second combined plot showing the grouped bar comparison for each fragment size
plt.figure(figsize=(16, 10))

# Get unique fragment sizes
fragment_sizes = sorted(pdr_df['Fragment Size (B)'].unique())
n_fragments = len(fragment_sizes)
n_sf = 4

# Set up subplot grid
fig, axes = plt.subplots(2, int(np.ceil(n_fragments/2)), figsize=(20, 12))
axes = axes.flatten() if n_fragments > 2 else [axes] if n_fragments == 1 else axes

for i, frag_size in enumerate(fragment_sizes):
    ax = axes[i]
    
    # Get data for this fragment size
    frag_data = pdr_df[pdr_df['Fragment Size (B)'] == frag_size]
    
    sf_values = []
    avg_pdrs = []
    
    for sf in [7, 8, 9, 10]:
        sf_row = frag_data[frag_data['SF'] == sf]
        if not sf_row.empty:
            sf_values.append(f'SF {sf}')
            avg_pdrs.append(sf_row['Avg (%)'].iloc[0])
    
    # Create bar chart for this fragment size
    bars = ax.bar(sf_values, avg_pdrs, color=[colors[sf] for sf in sf_values], 
                  alpha=0.8, edgecolor='black', linewidth=1.5)
    
    ax.set_title(f'Fragment Size: {frag_size}B', fontsize=14, fontweight='bold')
    ax.set_ylabel('Average PDR (%)', fontsize=12, fontweight='bold')
    ax.grid(True, alpha=0.3, axis='y', linestyle='--')
    
    # Add value labels on bars
    for bar, val in zip(bars, avg_pdrs):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'{val:.1f}%', ha='center', va='bottom', 
                fontsize=10, fontweight='bold')
    
    # Set consistent y-axis limits
    ax.set_ylim(94, 101)

# Hide unused subplots
for j in range(len(fragment_sizes), len(axes)):
    axes[j].set_visible(False)

plt.suptitle('PDR Comparison Across Different Fragment Sizes\nfor SF 7, 8, 9, and 10', 
             fontsize=18, fontweight='bold', y=0.98)
plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()
```
