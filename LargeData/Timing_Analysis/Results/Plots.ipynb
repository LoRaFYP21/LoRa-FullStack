# Timing Analysis

# SF = 7

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
df = pd.read_csv('TimingAnalysis.csv')

# Clean the data - remove commas from numeric values and convert to float
for col in ['Test 1 (ms)', 'Test 2 (ms)', 'Test 3 (ms)', 'Avg (ms)']:
    df[col] = df[col].str.replace(',', '').replace('X', np.nan)
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Filter data for SF 7
sf7_data = df[df['SF'] == 7].copy()

# Graph 1: Grouped Bar Chart - Transmission Times for Each Fragment Size (All Tests)
plt.figure(figsize=(12, 6))

# Set up data for grouped bars
fragment_sizes = sf7_data['Fragment Size (B)'].astype(str)
test1_times = sf7_data['Test 1 (ms)']
test2_times = sf7_data['Test 2 (ms)']
test3_times = sf7_data['Test 3 (ms)']
avg_times = sf7_data['Avg (ms)']

# Set the width of bars and positions
bar_width = 0.2
x_pos = np.arange(len(fragment_sizes))

# Create grouped bars
bars1 = plt.bar(x_pos - 1.5*bar_width, test1_times, bar_width, 
                label='Test 1', color='#E74C3C', alpha=0.8, edgecolor='black')
bars2 = plt.bar(x_pos - 0.5*bar_width, test2_times, bar_width, 
                label='Test 2', color='#3498DB', alpha=0.8, edgecolor='black')
bars3 = plt.bar(x_pos + 0.5*bar_width, test3_times, bar_width, 
                label='Test 3', color='#F39C12', alpha=0.8, edgecolor='black')
bars_avg = plt.bar(x_pos + 1.5*bar_width, avg_times, bar_width, 
                   label='Average', color='#27AE60', alpha=0.8, edgecolor='black', linewidth=2)

plt.xlabel('Fragment Size (B)', fontsize=14, fontweight='bold')
plt.ylabel('Transmission Time (ms)', fontsize=14, fontweight='bold')
plt.title('SF 7: Transmission Times by Fragment Size (All Tests)', fontsize=16, fontweight='bold')
plt.xticks(x_pos, fragment_sizes)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3, axis='y', linestyle='--')

# Add value labels on all bars
for bars in [bars1, bars2, bars3, bars_avg]:
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 f'{height:.1f}',
                 ha='center', va='bottom', fontsize=7, fontweight='bold')

plt.tight_layout()
plt.show()

# Graph 2: Line Chart - Variation of Average Transmission Time with Fragment Size
plt.figure(figsize=(10, 6))
plt.plot(sf7_data['Fragment Size (B)'], sf7_data['Avg (ms)'], 
         marker='o', linewidth=3, markersize=10, color='#2E86AB')

plt.xlabel('Fragment Size (B)', fontsize=14, fontweight='bold')
plt.ylabel('Average Transmission Time (ms)', fontsize=14, fontweight='bold')
plt.title('SF 7: Average Transmission Time vs Fragment Size', fontsize=16, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle='--')
plt.xticks(sf7_data['Fragment Size (B)'])

# Add value labels on points
for i, (x, y) in enumerate(zip(sf7_data['Fragment Size (B)'], sf7_data['Avg (ms)'])):
    plt.annotate(f'{y:.1f} ms', (x, y), textcoords="offset points", 
                xytext=(0,10), ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```
